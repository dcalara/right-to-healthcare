```python
import pandas as pd
```

# GDP_df

```python
# The path to our CSV file
file = "Current health expenditure (% of GDP).csv"

# Read our data into pandas
df = pd.read_csv(file, sep='delimiter', engine='python')
df
```

```python
GDP_df = df.drop(df.index[[0]])
GDP_df.head()
```

```python
header = GDP_df.iloc[0]
GDP_df = GDP_df.rename(columns = header)
GDP_df.head()
```

```python
# List columns
GDP_df.columns
```

```python
GDP_data = GDP_df.loc[:, ["Country Name","Country Code","Indicator Name","Indicator Code","1960", 
                       "2000","2001","2002","2003","2004","2005","2006","2007","2008","2009","2010",
                        "2011","2012","2013","2014","2015","2016","2017","2018"]]
GDP_data.head()
```
